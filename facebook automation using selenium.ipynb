```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

try:
    # Input data
    email = "your facebook  mail id"
    password = "facebook password"

    # Configure Chrome options to disable notifications
    chrome_options = Options()
    chrome_options.add_argument("--disable-notifications")

    # Latest Chromedriver is autodownloaded and loaded.
    driver = webdriver.Chrome(options=chrome_options, service=Service())

    # Facebook Login page
    driver.get("https://www.facebook.com/")
    print("Entering email")
    time.sleep(2)

    # sending email by name
    driver.find_element(By.NAME, 'email').send_keys(email)
    print("Entering password")
    time.sleep(2)

    # sending password by name
    driver.find_element(By.NAME, 'pass').send_keys(password)
    print("clicking login")
    time.sleep(2)

    # Defining wait for waiting until elements are clickable
    wait = WebDriverWait(driver, 10)

    # Clicking login button by NAME
    login_btn = wait.until(EC.element_to_be_clickable((By.NAME, 'login')))
    login_btn.click()

    # Wait until URL changes after login
    wait.until(EC.url_changes('https://m.facebook.com/'))

    # Clicking the "What's on your mind" field to start a post
    post_input = driver.find_element(By.XPATH, "//span[contains(text(), \"What's on your mind, Priyam?\")]")
    post_input.click()

    print("Writing post")

    # Writing the post content
    
    post_text_area = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.XPATH, "//div[@contenteditable='true']")))
    
    post_text_area.send_keys("hello this side priyam")
    print("Post content added")

    # Clicking the "Post" button to publish the post
    post_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//div[@aria-label='Post']")))
    post_button.click()

    print("Post button clicked. Waiting for post to be published...")

    # Wait for the success message or any indication that the post has been published
    time.sleep(5)  # Adjust the wait time as needed

    print("Post published successfully!")

    # Close the driver
    driver.quit()

except Exception as e:
    print(e)
```
